```python
def distance_matrix(cities):
  n = len(cities)
  distance_matrix = [[0 for _ in range(n)] for _ in range(n)]
  for i in range(n):
    for j in range(i + 1, n):
      x1, y1 = cities[i]
      x2, y2 = cities[j]
      distance = ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5
      distance_matrix[i][j] = distance_matrix[j][i] = distance
  return distance_matrix

def tsp_fitness(permutation, distance_matrix):
  total_distance = 0
  n = len(permutation)
  for i in range(n - 1):
    city1 = permutation[i]
    city2 = permutation[i + 1]
    total_distance += distance_matrix[city1][city2]
  total_distance += distance_matrix[permutation[n - 1]][permutation[0]]
  return total_distance

def selection(population, fitness_values):
  total_fitness = sum(fitness_values)
  probabilities = [fitness / total_fitness for fitness in fitness_values]
  selected_parents = []
  for _ in range(len(population) // 2):
    spin = random.random()
    probability_sum = 0
    for i in range(len(population)):
      probability_sum += probabilities[i]
      if probability_sum >= spin:
        selected_parents.append(population[i])
        break
  return selected_parents


def crossover(parent1, parent2):
  n = len(parent1)
  crossover_point = random.randint(1, n - 2)
  filtered_cities = [city for city in parent2 if city not in set(parent1[:crossover_point])]
  child1 = parent1[:crossover_point] + filtered_cities
  filtered_cities = [city for city in parent1 if city not in set(parent2[:crossover_point])]
  child2 = parent2[:crossover_point] + filtered_cities
  return child1, child2

def mutation(permutation, mutation_rate):
  n = len(permutation)
  if random.random() < mutation_rate:  # Perform mutation only with this probability
    i, j = random.sample(range(n), 2)
    permutation[i], permutation[j] = permutation[j], permutation[i]
  return permutation

def ga_tsp(distance_matrix, population_size, max_iterations, mutation_rate=0.01, crossover_prob=0.9):
    n = len(distance_matrix)
    population = [random.sample(range(n), n) for _ in range(population_size)]  # Randomly initialize population
    fitness_values = [tsp_fitness(permutation, distance_matrix) for permutation in population]

    for iteration in range(max_iterations):
        selected_parents = selection(population, fitness_values)
        children = []
        for i in range(0, len(selected_parents)-1, 2):
            if random.random() < crossover_prob:  # Perform crossover with this probability
                parent1, parent2 = selected_parents[i], selected_parents[i + 1]
                child1, child2 = crossover(parent1, parent2)
                children.append(child1)
                children.append(child2)
            else:
            # If not performing crossover, keep parents as they are
                children.append(selected_parents[i])
                children.append(selected_parents[i + 1])
            mutated_children = [mutation(child.copy(), mutation_rate) for child in children]
            children_fitness = [tsp_fitness(permutation, distance_matrix) for permutation in mutated_children]
            new_population = population + mutated_children
            new_fitness_values = fitness_values + children_fitness
            sorted_indices = sorted(range(len(new_population)), key=lambda i: new_fitness_values[i])
            population = [new_population[i] for i in sorted_indices[:population_size]]
            fitness_values = [new_fitness_values[i] for i in sorted_indices[:population_size]]

    best_permutation_index = fitness_values.index(min(fitness_values))
    best_permutation = population[best_permutation_index]
    best_value = fitness_values[best_permutation_index]
    return best_permutation, best_value


distance_matrix = distance_matrix(cities)
population_size = 200
max_iterations = 400

best_permutation, best_value = ga_tsp(distance_matrix, population_size, max_iterations)


print(f"Minimum found at: {best_permutation}")
print(f"Minimum value (distance): {best_value}")

```

```python
import random
import math
import pandas as pd
def distance(city1, city2):
  """Calculates the distance between two cities."""
  # Assuming coordinates are stored in city tuples (x, y)
  x1, y1 = city1
  x2, y2 = city2
  return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def total_distance(route, distances):
  """Calculates the total distance of a given route."""
  total = 0
  for i in range(len(route) - 1):
    city1 = route[i]
    city2 = route[i + 1]
    total += distances[city1][city2]
  total += distances[route[-1]][route[0]]  # Add distance back to starting city
  return total

def generate_initial_route(num_cities):
  """Generates a random initial route."""
  route = list(range(num_cities))
  random.shuffle(route)
  return route

def swap_cities(route, i, j):
  """Swaps two cities in the route."""
  route[i], route[j] = route[j], route[i]

def accept_worse(delta_distance, temperature):
  """Decides whether to accept a worse solution based on temperature."""
  if delta_distance < 0:
    return True
  p = math.exp(-delta_distance / temperature)
  return random.random() < p

def simulated_annealing(distances, starting_temp, cooling_rate, iterations):
  """Solves TSP using Simulated Annealing algorithm."""
  num_cities = len(distances)
  best_route = generate_initial_route(num_cities)
  best_distance = total_distance(best_route, distances)
  current_route = best_route.copy()
  current_distance = best_distance
  temperature = starting_temp
  c = 0
  for _ in range(iterations):
    for _ in range(num_cities):
      i, j = random.sample(range(num_cities), 2)  # Select two random cities
      swap_cities(current_route, i, j)
      new_distance = total_distance(current_route, distances)
      delta_distance = new_distance - current_distance

      if accept_worse(delta_distance, temperature):
        c+=1
        current_distance = new_distance
        if new_distance < best_distance:
          best_route = current_route.copy()
          best_distance = new_distance
      else:
        c=0
        swap_cities(current_route, i, j)  # Revert the swap if not accepted
      if c==3:
        temperature *= cooling_rate
        c=0

  return best_route, best_distance


distances = [[distance(city1, city2) for city2 in cities] for city1 in cities]
all_best_distances = []
for _ in range(30):
    temp = random.randint(5,100)
    cooling_rate = random.random()
    iterations = 100
    best_position, best_value = simulated_annealing(distances, temp, cooling_rate, iterations)
    print(f"Minimum value with temperature= {temp} , cooling_rate = {cooling_rate} , i = {iterations}  : {best_value}")
    all_best_distances.append(best_value)
plt.plot(range(len(all_best_distances)), all_best_distances, label="Best Distance")
plt.xlabel("Iteration")
plt.ylabel("Distance")
plt.title("Best Distance per Iteration -sa")
plt.legend()
plt.show()
print(f"Minimum found at: {best_position}")
print(f"Minimum value: {best_value}")
```
